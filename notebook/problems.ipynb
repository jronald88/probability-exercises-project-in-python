## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

def die_roll(no_rolls = 1000):
    # 
    count = 0
    #rolls_A =[]
    for i in range(no_rolls):
        die_A = np.random.randint(1,7)
        die_B = random.randint(1,6)
        sum = die_A + die_B

        if (sum > 7) or (sum % 2 == 0):
            count +=1
        
    return(count/no_rolls)
print(f'The probability of rolling a number divisible by 2 or greater than 7 is {die_roll()}')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
count_1 = 0
count_2 = 0
for j in range(1000):
        
    white_count = 0
    red_count = 0
    green_count = 0
    
    for k in range(5):
        colour = ball_box[np.random.randint(0,60)]
        if colour == 'Red':
            red_count += 1
        elif colour == 'White':
            white_count += 1
        else:
            green_count += 1
    if (red_count == 3) and (white_count == 2):
        count_1 += 1
    if (red_count == 5) or (white_count == 5) or (green_count == 5):
        count_2 += 1

sol_1 = count_1/1000*100
sol_2 = count_2/1000*100
print(f'The probability of selecting 3 White and 2 Red counters is %.2f' % sol_1,"%")    
print(f'The probability of selecting 5 colours the same is %.2f' % sol_2,"%")    

pd.DataFrame([sol_1,sol_2,100-sol_1-sol_2]).plot.pie(subplots=True)

```
